# **1) Introduction to matplotlib**
---


---
### ***1.1 What is matplotlib?***

▶ is a powerful Python library for creating static, interactive, and animated visualizations. It is widely used for data visualization in scientific computing, machine learning, and data analysis.

- Developed by John D. Hunter in 2003, Matplotlib provides a MATLAB-like interface and is highly customizable

- **matplotlib:**

  - **"Mat" :** Referring to MATLAB, as Matplotlib was originally designed to replicate MATLAB's plotting capabilities in Python.

  - **"Plot" :** Emphasizing its core functionality for creating plots and visualizations.

  - **"Lib" :** Short for Library.
---

---
### ***1.2 Key Characteristics of Matplotlib***

- **Highly Customizable**:
  - Supports extensive customization of plots, labels, colors, fonts, and styles.

  - Allows fine-tuning of axes, grids, legends, and annotations.

- **Multi-Platform & Backend Support**

  - Works with Python scripts, Jupyter notebooks, and web applications.

  - Supports different backends (e.g., Qt, GTK, Tkinter, Web).

- **Wide Range of Plot Types**
  - Basic Plots: Line plots, bar charts, histograms, scatter plots.

  - Advanced Plots: Contour plots, 3D plots, polar plots, heatmaps.

  - Specialized Plots: Box plots, violin plots, pie charts, error bars.

- **Integration with Other Libraries**
  - Works seamlessly with NumPy, Pandas, Seaborn, and SciPy.

  - Can be used alongside Jupyter Notebooks for interactive plotting.
- **Object-Oriented & Pyplot Interface**
  - Object-Oriented (OO) Approach: Full control over figure and axes.

  - Pyplot Interface: MATLAB-like syntax for quick plotting.

- **Export & Save Options**
  - Supports saving plots in PNG, PDF, SVG, EPS, and JPEG formats.

### ***1.6 Matplotlib Architecture (Figure, Axes, Artist, Renderer)***

```python
fig, ax = plt.subplots()
```
▶ **Definition** – Matplotlib’s architecture is based on a hierarchy:
- **Figure:** The whole window or page where everything is drawn.
- **Axes:** The part of the figure where data is plotted (like a subplot).
- **Artist:** Everything you see on a plot (titles, lines, labels, etc.)
- **Renderer:** Handles converting artists to visual output.

💠 **Usage:**  
Understanding this structure helps build complex plots by customizing each layer.

````python
# EX1:
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot([1, 2], [3, 4])
````

````python
#EX2:
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
axs[0, 0].plot([1, 2], [1, 4], 'r')
axs[0, 1].bar([1, 2, 3], [3, 5, 7])
axs[1, 0].hist([1, 2, 1, 1, 2, 3])
axs[1, 1].scatter([1, 2, 3], [3, 2, 1])
fig.suptitle("Multiple Subplots Example")
plt.tight_layout()
plt.show()
````

### **A) pyplot Functions**

- **'plt.plot()'**

```python
plt.plot(x, y, format_string, **kwargs)
```
- Parameters:
  - **x:** array-like, x-coordinates
  - **y:** array-like, y-coordinates  
  - **format_string:** string, color/marker/line style (e.g., 'bo-')
  - **kwargs:** additional styling options (linewidth, alpha, label, etc.)

▶ **Definition** – The plot() function creates line plots and scatter plots.
- It's the most fundamental plotting function in matplotlib, used to visualize relationships between variables with lines and markers.

💠 **Usage:**
   - Creating line graphs to show trends over time
   - Plotting mathematical functions and relationships
   - Displaying multiple data series on the same plot
   - Basic data visualization and exploratory analysis

````python
# Example 1:
import numpy as np

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x,y)

plt.title("Basic Line Blot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
````

````python
# Example 2:

# Multiple lines with different styles
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x/2)

plt.figure(figsize=(12, 8))
plt.plot(x, y1, 'b-', linewidth=2, label='sin(x)', alpha=0.8)
plt.plot(x, y2, 'r--', linewidth=2, label='cos(x)', alpha=0.8)
plt.plot(x, y3, 'g:', linewidth=3, label='tan(x/2)', alpha=0.8)

plt.title('Trigonometric Functions', fontsize=16)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.xlim(0, 10)
plt.ylim(-2, 2)
plt.show()
````

- **Parameters:**

▶ **Definition** – The scatter() function creates scatter plots to visualize the relationship between two continuous variables.

💠 **Usage:**

````python
# Advanced scatter plot with size and color coding
n = 200
x = np.random.randn(n)
y = np.random.randn(n)
colors = np.random.rand(n)
sizes = 1000 * np.random.rand(n)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(scatter, label='Color Scale')
plt.title('Advanced Scatter Plot with Size and Color Coding', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# Grouped bar chart
categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [20, 35, 30, 25]
product_b = [25, 30, 35, 30]
product_c = [15, 25, 25, 20]

x = np.arange(len(categories))
width = 0.25

plt.figure(figsize=(10, 6))
plt.bar(x - width, product_a, width, label='Product A', color='skyblue')
plt.bar(x, product_b, width, label='Product B', color='orange')
plt.bar(x + width, product_c, width, label='Product C', color='lightgreen')

plt.title('Quarterly Sales Comparison', fontsize=14)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Sales (in thousands)', fontsize=12)
plt.xticks(x, categories)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')
plt.show()
````

### **B) Stateful Interface Management**

- **`Stateful Interface`**

```python
plt.plot([1, 2, 3], [1, 4, 9])  # Creates figure and axes automatically
plt.title('Current Figure')     # Applies to current axes
plt.xlabel('X')                 # Applies to current axes
plt.show()                      # Shows current figure
```

▶ **Definition** – The stateful interface automatically manages the current figure and axes, allowing you to create plots without explicitly creating figure objects. Matplotlib keeps track of the "current" figure and axes.

💠 **Usage:**
   - Quick and simple plotting for exploration
   - Interactive data analysis
   - Learning matplotlib basics
   - Creating single plots or simple multi-plot layouts

````python
# First subplot
plt.subplot(2, 2, 1)
plt.plot(np.random.randn(100).cumsum(), 'b-')
plt.title('Random Walk')
plt.grid(True)

# Second subplot
plt.subplot(2, 2, 2)
plt.hist(np.random.randn(1000), bins=30, alpha=0.7)
plt.title('Histogram')

# Third subplot
plt.subplot(2, 2, 3)
plt.scatter(np.random.randn(100), np.random.randn(100), alpha=0.6)
plt.title('Scatter Plot')

# Fourth subplot
plt.subplot(2, 2, 4)
data = np.random.rand(5, 5)
plt.imshow(data, cmap='hot')
plt.title('Heatmap')
plt.colorbar()

plt.tight_layout()
plt.show()
````

- **`plt.gca()` and `plt.gcf()`**

```python
# Get current axes and figure
ax = plt.gca()  # Get Current Axes
fig = plt.gcf()  # Get Current Figure

# Use returned objects
ax.set_title('Current Axes')
fig.suptitle('Current Figure')
```

▶ **Definition** – gca() returns the current axes object, and gcf() returns the current figure object. These functions bridge stateful and object-oriented interfaces.

💠 **Usage:**
   - Accessing current axes/figure for detailed customization
   - Bridging stateful and object-oriented approaches
   - Applying complex formatting to current plots
   - Advanced plot customization

````python
fig = plt.figure(figsize=(15, 10))
fig.suptitle('Advanced Figure Layout', fontsize=16)

# Create subplots with different sizes
ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 3, (4, 5))  # Span two columns
ax5 = plt.subplot(2, 3, 6)

# Plot data in each subplot
x = np.linspace(0, 10, 100)
axes = [ax1, ax2, ax3, ax4, ax5]
functions = [np.sin, np.cos, np.tan, lambda x: np.sin(x) * np.cos(x), np.exp]
titles = ['sin(x)', 'cos(x)', 'tan(x)', 'sin(x)*cos(x)', 'exp(x)']

for ax, func, title in zip(axes, functions, titles):
    if title == 'exp(x)':
        y = func(-x/2)  # Avoid overflow
    elif title == 'tan(x)':
        y = func(x)
        y = np.clip(y, -5, 5)  # Clip extreme values
    else:
        y = func(x)

    ax.plot(x, y, linewidth=2)
    ax.set_title(title)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Using gca() and gcf()
plt.figure(figsize=(8, 6))
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

# Get current axes and figure
current_ax = plt.gca()
current_fig = plt.gcf()

# Customize using object-oriented approach
current_ax.set_title('Using gca() and gcf()', fontsize=14)
current_ax.set_xlabel('X values')
current_ax.set_ylabel('Y values')
current_fig.patch.set_facecolor('lightgray')

plt.show()
````

### **C) Artist Hierarchy**

fig, ax = plt.subplots()

# Artist properties (common to all artists)
artist.set_alpha(0.5)        # Transparency

````python
import numpy as np
from matplotlib.lines import Line2D

# Create figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create Line2D objects directly
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create custom Line2D artists
line1 = Line2D(x, y1, linewidth=3, color='blue', linestyle='-',
               marker='o', markersize=5, markevery=10, label='sin(x)')
line2 = Line2D(x, y2, linewidth=2, color='red', linestyle='--',
               marker='s', markersize=4, markevery=15, label='cos(x)')

# Add lines to axes
ax.add_line(line1)
ax.add_line(line2)

# Set axes properties
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)
ax.set_title('Custom Line2D Artists', fontsize=14)
ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()
````
