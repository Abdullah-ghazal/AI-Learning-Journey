# **1) Introduction to seaborn**
---


---
### ***1.1 What is seaborn?***

▶ Seaborn is a powerful Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

---

--
### ***1.2 Overview***

- Seaborn enhances the default Matplotlib styles and integrates tightly with Pandas DataFrames.
- It simplifies the process of creating complex plots and includes built-in themes, color palettes, and functions for visualizing univariate and bivariate distributions, categorical data, regression relationships, and more.

---

### ***1.6 Seaborn vs. Matplotlib***

| Feature            | Seaborn                             | Matplotlib                          |
|--------------------|-------------------------------------|-------------------------------------|
| **Interface**          | High-level                          | Low-level                           |
| **Data Structures**  | Pandas DataFrames                   | Arrays, lists                       |
| **Built-in Themes**   | Yes                                 | Limited                             |
| **Statistical Plots**  | Built-in                            | Requires manual implementation      |
| **Color Palettes**     | Simplified and aesthetically(جماليًا) better | Basic                               |

- **Use Seaborn** if you want:
  - Fast and beautiful statistical plots.
  - Tight Pandas integration.
  - Built-in themes and color palettes.

- **Use Matplotlib** if you want:
  - Fine-grained control over every element of the plot.
  - Custom or non-statistical plots.

# **2) Figure-Level vs. Axes-Level Functions**
---


---
### ***2.1 OverView:***
- At its core, Matplotlib (and by extension, Seaborn) operates on two main components:

    - **Figure:**
      - This is the entire window or page on which everything is drawn.
      - Think of it as the canvas.

    - **Axes:**
      - This is where the data is actually plotted.
      - It's the region of the image where the data appears, and it can contain multiple plots (subplots).
      - A Figure can contain multiple Axes.
---

---

### ***2.2 Figure-Level Functions***

▶ Figure-level functions create an entire **Figure**, including one or more subplots, and manage the layout and overall appearance.

💠 Useful for **faceting** (plotting multiple subplots using `col`, `row`, or `hue`), adjusting overall plot size, and managing subplot layouts.

- **`sns.relplot()`** :  General-purpose relational plots (scatter, line).
- **`sns.displot()`** : General-purpose distribution plots (histogram, KDE , ECDF).
- **`sns.catplot()`** : General-purpose categorical plots (box, violin, bar, etc.).
- **`sns.lmplot()`** : Plots regression models (scatter with regression line).

```python
sns.relplot(data=df, x="x", y="y", hue="category", kind="scatter", col="gender")
```
- **Common parameters:**
  * **data:** The DataFrame containing the data.
  * **x, y:** Column names for x and y axes.
  * **col**, **row:** Column names to create subplots based on categories in these columns.
  * **hue:** Column name to differentiate data points by color.
  * **kind:** Specifies the type of plot (e.g., "scatter", "line" for relplot; "hist", "kde" for displot; "box", "violin" for catplot).
  * **col_wrap:** Number of columns to wrap facets.
  * **height**, **aspect**: Control the size of each facet.

---

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set a style for better aesthetics
sns.set_theme(style="whitegrid")

# Load example datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
````

````python
# Create a grid of scatter plots using a Figure-level function
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="time", col="day",
                kind="scatter", height=4, aspect=1.2)
g.set_axis_labels("Total Bill ($)", "Tip ($)")
g.set_titles("Day: {col_name}")
plt.suptitle("Tip vs. Total Bill by Day (Figure-level)", y=1.02) # Adjust suptitle position
plt.show()
````

---

### ***2.3 Axes-Level Functions***

▶ Axes-level functions draw **single plots** on a specific `matplotlib.axes.Axes` object.
- They do not manage the figure layout or size.

💠 Best for creating individual plots and when embedding plots into custom Matplotlib figures.

- **`scatterplot()`** : Draws scatter plots.
- **`lineplot()`** : Draws line plots.
- **`barplot()`** : Draws bar plots (mean/confidence interval).
- **`countplot()`** : Draws bar plots of counts.
- **`boxplot()`** : Draws box plots.
- **`violinplot()`** : Draws violin plots.
- **`stripplot()`** : Draws scatter plots with categories.
- **`swarmplot()`** : Draws categorical scatter plots without overlap.
- **`pointplot()`** : Draws point plots (mean/confidence interval).
- **`histplot()`** : Draws histograms.
- **`kdeplot()`** : Draws Kernel Density Estimate plots.
- **`ecdfplot()`** : Draws Empirical Cumulative Distribution Function plots.
- **`jointplot()`** : Creates a JointGrid; acts like an Axes-level function for the main plot area but returns a `JointGrid` object.
- **`pairplot()`** : Creates a PairGrid to show pairwise relationships in a dataset.

```python
sns.relplot(data=df, x="x", y="y", hue="category", kind="scatter", col="gender")
```

- **Common Parameters:**
  - **data**:  The `DataFrame` containing the data.

  - **x**, **y**:  Column names in `data` for the x and y axes.

  - **hue**:  Column name used to group and color data points by category.

  - **ax**:  A `Matplotlib Axes` object to draw the plot on. (If not provided, Seaborn will create a new Axes.)

````python
# Create a single scatter plot using an Axes-level function
plt.figure(figsize=(8, 6)) # Create a figure explicitly
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Tip vs. Total Bill by Time of Day (Axes-level)")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
````

---

### ***2.4 Differences and Use Cases***

| Feature                    | Figure-Level Functions                   | Axes-Level Functions                        |
|----------------------------|------------------------------------------|---------------------------------------------|
| **Scope**                     | Whole figure                             | Single axes                                 |
| **Suitable for**              | Complex layouts, faceting                | Single plots, fine customization            |
| **Custom layout**             | Supports `col`, `row`, `hue` faceting    | No faceting                                 |
| **Return object**             | `FacetGrid`                              | `Axes`                                      |
| **Matplotlib integration**    | Moderate                                 | High – direct control over the Axes         |

---

````python
# Create a custom figure with multiple subplots and use Axes-level functions
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True) # Share y-axis for comparison

# Plot 1: Scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", ax=axes[0])
axes[0].set_title("Total Bill vs. Tip (Smoker)")
axes[0].set_xlabel("Total Bill ($)")
axes[0].set_ylabel("Tip ($)")

# Plot 2: Bar plot of tip by day
sns.barplot(data=tips, x="day", y="tip", hue="time", ax=axes[1], errorbar=None) # errorbar=None removes confidence intervals
axes[1].set_title("Average Tip by Day and Time")
axes[1].set_xlabel("Day of Week")
axes[1].set_ylabel("Average Tip ($)")

# Plot 3: Distribution of total bill
sns.histplot(data=tips, x="total_bill", kde=True, ax=axes[2], bins=15)
axes[2].set_title("Distribution of Total Bill")
axes[2].set_xlabel("Total Bill ($)")
axes[2].set_ylabel("Count")

plt.tight_layout() # Adjust layout to prevent overlap
plt.suptitle("Custom Multi-Panel Plot with Axes-level Functions", y=1.02, fontsize=16)
plt.show()
````

---

### ***2.5 Best Practices***

- Use **Axes-level** for small/simple plots or tight Matplotlib integration.
- Use **Figure-level** when working with **faceting** or needing more layout control.
- Always set titles, axis labels, and legends explicitly for readability.
- Prefer Seaborn’s object-oriented approach for cleaner, more manageable code.

# **3) Categorical Plots**
---

---


---
### ***3.1 OverView:***

▶ Seaborn offers a variety of plots for visualizing categorical variables.
 - These plots are useful for comparing groups, understanding distributions within categories, and highlighting relationships between categorical and numerical data.

---

---

### ***3.2 Axes-Level Categorical Plots***

### **`barplot()`**
Shows the **point estimate** (default: mean) and **confidence interval** for a quantitative variable across categories.

```python
sns.barplot(data=df, x="category", y="value", estimator=np.mean, errorbar=("ci", 95), capsize=0.1)
```

- **Parameters**:
  - `x`, `y`: categorical, quantitative
  - `hue`: subgrouping variable
  - `estimator`: function (default `np.mean`)
  - `errorbar`: tuple or None (e.g., `("ci", 95)`)
  - `capsize`: length of error bar caps

✅ **Use for** comparing mean/median across groups with statistical confidence.

---

### **`countplot()`**
Bar plot showing the **count** of observations in each categorical group.

```python
sns.countplot(data=df, x="category", hue="group")
```

- **Parameters**:
  - `x` or `y`: categorical
  - `hue`: subgrouping variable
  - `data`: DataFrame

✅ **Use for** frequency counts of categories.

---

### **`boxplot()`**
Displays **five-number summary** (min, Q1, median, Q3, max) and **outliers** for quantitative data across categories.

```python
sns.boxplot(data=df, x="group", y="score", whis=1.5, sym="o")
```

- **Parameters**:
  - `x`, `y`: categorical, quantitative
  - `hue`: grouping
  - `whis`: whisker range
  - `sym`: outlier symbol (e.g., `"o"` or `""`)

✅ **Use for** summarizing distribution and spotting outliers.

---

### **`violinplot()`**
Combines KDE (distribution) with boxplot for a richer view of data spread and density.

```python
sns.violinplot(data=df, x="group", y="score", inner="box", split=False)
```

- **Parameters**:
  - `x`, `y`: categorical, quantitative
  - `hue`: grouping
  - `inner`: "box", "quartile", "point", or None
  - `split`: split violins if `hue` is set

✅ **Use for** showing distribution shape and summary.

---

### **`stripplot()`**
Scatter plot for categorical axis. Good for small datasets. Points can **overlap**.

```python
sns.stripplot(data=df, x="group", y="value", jitter=True)
```

- **Parameters**:
  - `x`, `y`: categorical, quantitative
  - `jitter`: scatter points along category axis
  - `hue`: grouping

✅ **Use for** visualizing raw data points.

---

### **`swarmplot()`**
Like stripplot, but prevents overlapping by adjusting point positions.

```python
sns.swarmplot(data=df, x="group", y="value", dodge=True)
```

- **Parameters**:
  - `x`, `y`: categorical, quantitative
  - `hue`: grouping
  - `dodge`: separate hue groups side-by-side

✅ **Use for** visualizing all values without overlap.

---

### **`pointplot()`**
Shows **point estimates** (mean or other) and **confidence intervals** as lines and points.

```python
sns.pointplot(data=df, x="time", y="score", hue="group", estimator=np.mean, capsize=0.2)
```

- **Parameters**:
  - `x`, `y`: categorical, quantitative
  - `hue`: grouping
  - `estimator`: statistical function (e.g., `np.mean`, `np.median`)
  - `capsize`: length of cap lines on error bars

✅ **Use for** showing trends or interactions between groups.

---
---

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")

plt.figure(figsize=(10, 6))

# Bar plot: Average tip by day
plt.subplot(2, 2, 1) # Create a 2x2 grid, first subplot
sns.barplot(data=tips, x="day", y="tip", errorbar=("ci", 95))
plt.title("Average Tip by Day")
plt.ylabel("Average Tip ($)")

# Count plot: Number of observations per day
plt.subplot(2, 2, 2)
sns.countplot(data=tips, x="day", hue="smoker")
plt.title("Number of Patrons by Day (Smoker Status)")
plt.ylabel("Count")

# Box plot: Distribution of total bill by day
plt.subplot(2, 2, 3)
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Total Bill Distribution by Day")
plt.ylabel("Total Bill ($)")

# Violin plot: Distribution of total bill by day with hue
plt.subplot(2, 2, 4)
sns.violinplot(data=tips, x="day", y="total_bill", hue="time", split=True, inner="quartile")
plt.title("Total Bill Distribution by Day and Time")
plt.ylabel("Total Bill ($)")

plt.tight_layout()
plt.show()
````

---
### ***3.3 Figure-Level Categorical Plot***

- **`catplot()`** – Combines all above plots with `kind` parameter.

```python
sns.catplot(data=df, x="category", y="value", kind="violin", col="gender")
```

✅ `catplot()` is useful when faceting is needed (`col`, `row`, `hue`).

---

````python
# Use catplot to create a grid of box plots
g = sns.catplot(data=tips, x="day", y="total_bill", col="time",
                kind="box", height=4, aspect=0.9)
g.set_axis_labels("Day of Week", "Total Bill ($)")
g.set_titles("Time: {col_name}")
plt.suptitle("Total Bill Distribution by Day and Time (using catplot)", y=1.02)
plt.show()
````

---

```python
df["group"] = df["group"].astype("category")
```

# **4) Distribution Plots**
---

---
### ***4.1 OverView:***
- Distribution plots are fundamental for understanding the underlying patterns and shapes of your data.
- They show how values are distributed, identify central tendencies, spread, and potential outliers.

💠 **Distribution plots are used to:**

  - **Visualize data frequency**: How often certain values or ranges of values appear.
  - **Identify shape**: Whether the data is symmetric, skewed, unimodal, multimodal, etc.
  - **Detect outliers**: Values that fall far from the main cluster of data.
  - **Compare distributions**: See how the distribution of a variable differs across groups or under different conditions.

- **They are crucial for:**
  - Assessing assumptions in statistical modeling
  - Understanding variability
  - Gaining initial insights into the data’s characteristics

---

---

### ***4.2 Univariate Distribution Plots (Axes-Level)***

- #### **`histplot()`**

```python
sns.histplot(data=df, x="value", bins=30, kde=True, stat="density", hue="group")
```

- **Parameters**:
  - `x` or `y`: Variable to plot
  - `bins`: Number of bins
  - `binwidth`: Width of each bin
  - `kde`: Overlay KDE curve (True/False)
  - `stat`: "count", "frequency", "density", "probability"
  - `hue`: Grouping variable for color split

✅ Great for understanding raw data distribution

---

- #### **`kdeplot()`**

```python
sns.kdeplot(data=df, x="value", bw_adjust=0.8, fill=True, cut=2)
```

- **Parameters**:
  - `bw_adjust`: Bandwidth adjustment (smoothing)
  - `fill`: Fill the area under the curve
  - `cut`: Extend KDE beyond data range
  - `hue`: Group-wise KDE

✅ Best for smoothed, continuous distribution curves

---

- #### **`ecdfplot()`**

```python
sns.ecdfplot(data=df, x="value", hue="group", stat="proportion")
```

- **Parameters**:
  - `stat`: "count" or "proportion"
  - `hue`: For multiple ECDFs by group

✅ Ideal for comparing value accumulation over range

---

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")

plt.figure(figsize=(15, 5))

# Histplot of total bill
plt.subplot(1, 3, 1)
sns.histplot(data=tips, x="total_bill", kde=True, bins=15)
plt.title("Distribution of Total Bill (Histogram with KDE)")
plt.xlabel("Total Bill ($)")
plt.ylabel("Count")

# KDE plot of tip
plt.subplot(1, 3, 2)
sns.kdeplot(data=tips, x="tip", fill=True, bw_adjust=1.5) # Smoother KDE
plt.title("Distribution of Tip (KDE)")
plt.xlabel("Tip ($)")
plt.ylabel("Density")

# ECDF plot of total bill by time
plt.subplot(1, 3, 3)
sns.ecdfplot(data=tips, x="total_bill", hue="time")
plt.title("ECDF of Total Bill by Time")
plt.xlabel("Total Bill ($)")
plt.ylabel("Proportion")

plt.tight_layout()
plt.show()
````

---

### ***4.3 Bivariate Distribution Plots (Axes/Grid-Level)***

- #### **`jointplot()`**

```python
sns.jointplot(data=df, x="height", y="weight", kind="kde", height=6, ratio=4, space=0.2)
```

- **Parameters**:
  - `kind`: "scatter", "kde", "hist", "hex", "reg", "resid"
  - `height`, `ratio`, `space`: Layout controls

✅ Combines two-variable relationship + univariate margins

---

- #### **`pairplot()`**

```python
sns.pairplot(data=df, hue="species", kind="scatter", diag_kind="kde", corner=True)
```

- **Parameters**:
  - `vars`: Select columns
  - `hue`: Group variable
  - `kind`: Type of off-diagonal plot
  - `diag_kind`: "kde" or "hist"
  - `corner`: Whether to only plot lower triangle

✅ Great for full data overview

---

````python
# Jointplot: Scatter plot with marginal histograms
g = sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", height=7, ratio=4)
g.set_axis_labels("Total Bill ($)", "Tip ($)")
g.fig.suptitle("Total Bill vs. Tip (Jointplot - Scatter)", y=1.02)
plt.show()

# Jointplot: KDE with marginal KDEs
g = sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde", height=7, fill=True)
g.set_axis_labels("Total Bill ($)", "Tip ($)")
g.fig.suptitle("Total Bill vs. Tip (Jointplot - KDE)", y=1.02)
plt.show()

# Pairplot: Pairwise relationships for numerical variables
# Only select numerical columns for pairplot to avoid errors with non-numeric
numerical_tips = tips[['total_bill', 'tip', 'size']]
sns.pairplot(numerical_tips, diag_kind="kde", kind="scatter")
plt.suptitle("Pairwise Relationships of Numerical Tip Data", y=1.02)
plt.show()
````

---

### ***4.4 Figure-Level Distribution Function***

#### - **`displot()`**

```python
sns.displot(data=df, x="value", hue="group", kind="kde", col="category", fill=True)
```

- **Parameters**:
  - `kind`: "hist", "kde", or "ecdf"
  - `col`, `row`, `hue`: For faceting
  - `col_wrap`, `height`, `aspect`: Layout

✅ Ideal for large multi-group comparisons

---

````python
# Displot: Faceted histograms
g = sns.displot(data=tips, x="total_bill", col="time", row="smoker",
                kind="hist", bins=10, height=3, aspect=1.2)
g.set_axis_labels("Total Bill ($)", "Count")
g.set_titles(col_template="Time: {col_name}", row_template="Smoker: {row_name}")
plt.suptitle("Total Bill Distribution by Time and Smoker (using displot)", y=1.02)
plt.show()
````

---

### ***4.5 Kernel Density Estimation (KDE) Settings***

▶ KDE estimates the probability density function by smoothing data.

- **`bw_adjust`**:
  - `0.5`: More sensitive, detailed
  - `2`: Smoother, more general

```python
sns.kdeplot(data=df["value"], bw_adjust=0.5)
```

- **`cut`**: Extends KDE beyond data limits (default = 3)
- **`fill`**: Fills area under curve

---

---

### ***4.6 Histogram Binning (bins, binwidth)***

- **`bins`**:
  - Number of bins (e.g., `bins=20`)
  - Sequence of bin edges
  - `"auto"`: Auto-selected

- **`binwidth`**:
  - Width of each bin

```python
sns.histplot(data=df, x="value", binwidth=5)
```

-  **Best Practice**:
  - Too few bins → over-simplification
  - Too many bins → too noisy

---

````python
# Displot: Faceted KDE with custom bandwidth and hue
g = sns.displot(data=tips, x="tip", col="day", hue="smoker",
                kind="kde", fill=True, height=4, aspect=1,
                bw_adjust=0.8, palette="viridis",
                facet_kws=dict(margin_titles=True))
g.set_axis_labels("Tip ($)", "Density")
g.set_titles(col_template="{col_name} Day")
plt.suptitle("Tip Distribution by Day and Smoker (displot with Custom KDE)", y=1.02)
plt.show()

# Histplot with custom bins and stat
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x="total_bill", hue="time", multiple="stack",
             bins=[0, 10, 20, 30, 40, 50], stat="density", palette="muted")
plt.title("Total Bill Distribution by Time (Stacked Density Histogram)")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.show()
````
