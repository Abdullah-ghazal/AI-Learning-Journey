### ***1.1 What is Pandas?***
- **Pandas = Panel data**
> is a powerful open-source Python library used for data manipulation, analysis, and cleaning. It provides high-level data structures and functions designed to make working with structured (tabular, multidimensional, time series) data easy and intuitive.

  - Load and analyze datasets (CSV, Excel, SQL, JSON, etc.)

  - Clean or transform data (missing values, duplicates, formatting)

  - Perform statistical or mathematical operations

  - Group, filter, and summarize data

  - Merge, join, or reshape datasets

  - Work with time-series data

## **2.1 Series**

#### **a) Creation:**
```python
pd.Series(data=None, index=None, dtype=None, name=None, copy=False)
```
â–¶ **Definition** â€“ A Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). It's like a column in a spreadsheet or a single column from a DataFrame.

ðŸ’  **Usage:** Use Series when you need to work with a single column of data or when you want to perform operations on a one-dimensional dataset with labeled indices.

````python
# Create series from list:
numbers = pd.Series([1, 2, 3, 4])
print(numbers)
````

````python
# Create series with custom indices
fruits = pd.Series(['Apple', 'Banana', 'Orange'], index=['a', 'b', 'c'])
print(fruits)
````

````python
# Create Series from dictionary with mixed datatypes
data = {'name' : 'Abdullah', 'age': 22, 'salary': 0.0, 'married': False}
person_data = pd.Series(data)

print(person_data)
````

````python
# Create Series with datetime index
dates = pd.date_range('2024-01-01', periods=5, freq='D')
sales = pd.Series([100, 150, 200, 175, 300], index=dates, name='Daily Sales')
print(sales)
````

- ### **`dropna()`**

```python

```
â–¶ **Definition** â€“ Returns a Series with missing values removed.

ðŸ’  **Usage:** Use to clean your data by removing rows with missing values.

````python
data = pd.Series([1, 2, None, 4, None, 6])
clean_data = data.dropna()
print(clean_data)

# remove row = 2, 4
````

- ### **`fillna()`**

```python
series.fillna(value=None, method=None, inplace=False)
```
â–¶ **Definition** â€“ Fills missing values with a specified value or method.

ðŸ’  **Usage:** Use to replace missing values with meaningful substitutes.

````python
data = pd.Series([1, 2, None, 4])
filled_data = data.fillna(99)  # Replace NaN with 99
print(filled_data)
````

````python
# Forward fill
# fills missing (NaN) values by propagating the last valid (non-null) value forward.
filled_forward = data.fillna(method='ffill')
print(filled_forward)
````

- ### **`astype()`**

```python

```

````python
numbers_str = pd.Series(['1', '2', '3'])
numbers_int = numbers_str.astype(int)
print(numbers_int.dtype)
````

- ### **`map()`**

```python

```
â–¶ **Definition** â€“ Maps values of Series according to input correspondence (dict, Series, or function).

ðŸ’  **Usage:** Use to transform values based on a mapping or function.

````python
sizes = pd.Series(['S', 'M', 'L', 'S', 'L'])
size_map = {'S': 'Small', 'M': 'Medium', 'L': 'Large'}
full_sizes = sizes.map(size_map)
print(full_sizes)
````

- ### **`apply()`**

```python

```
â–¶ **Definition** â€“ Applies a function along the Series.

ðŸ’  **Usage:** Use to apply custom functions to transform your data.

````python
import math

numbers = pd.Series([1, 2, 3, 4, 5])
squared = numbers.apply(lambda x: x**2)
print(squared)


sqrt_numbers = numbers.apply(math.sqrt)
print(sqrt_numbers)
````

- ### **`copy()`**

```python

```

````python
original = pd.Series([1, 2, 3])
copy_series = original.copy()
copy_series[0] = 999
print(original)
print(copy_series)
````

## **2.2 DataFrame**

#### **a) Creation:**

```python
pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
```
â–¶ **Definition** â€“ A DataFrame is a 2-dimensional labeled data structure with columns of potentially different types.
- It's like a spreadsheet or SQL table, or a dictionary of Series objects.

ðŸ’  **Usage:** Use DataFrame when you need to work with tabular data with multiple columns and rows, perform data analysis, or manipulate structured datasets.

````python
# Create DataFrame from dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Paris', 'London']}
df = pd.DataFrame(data)
print(df)
````

````python
# Create DataFrame from list of dictionaries
employees = [
    {'name': 'Alice', 'department': 'Engineering', 'salary': 90000, 'years': 3},
    {'name': 'Bob', 'department': 'Marketing', 'salary': 70000, 'years': 5},
    {'name': 'Charlie', 'department': 'Engineering', 'salary': 95000, 'years': 2}
]
df = pd.DataFrame(employees)
print(df)
````

````python
# Create DataFrame with custom index
df_custom = pd.DataFrame(
    data=[[1, 2, 3], [4, 5, 6], [7, 8, 9]],
    index=['row1', 'row2', 'row3'],
    columns=['A', 'B', 'C']
)

print(df_custom)
````

## **2.3 Index Objects**
---

#### **a) Types**

- **`Index`**

```python
pd.Index(data=None, dtype=None, name=None)
```
â–¶ **Definition**
  - The basic index object for pandas objects.
  - Immutable array implementing an ordered, sliceable set.

ðŸ’  **Usage:** Use as a general-purpose index for any data type.

````python
idx = pd.Index(['a', 'b', 'c', 'd'])
df = pd.DataFrame({'values': [1, 2, 3, 4]}, index=idx)
print(df)
````

- **`DatetimeIndex`**

```python
pd.DatetimeIndex(data=None, freq=None, tz=None, name=None)
```
â–¶ **Definition** â€“ Index for datetime values, enabling time-series functionality.

ðŸ’  **Usage:** Use for time-series data and date-based operations.

````python
dates = pd.date_range('2024-01-01', periods=5, freq='D')
df = pd.DataFrame({'sales': [100, 150, 200, 175, 300]}, index=dates)
print(df)
# Enables time-series operations like resampling
````

- **`TimedeltaIndex`**

```python
pd.TimedeltaIndex(data=None, unit=None, freq=None, name=None)
```
â–¶ **Definition** â€“ Index for timedelta values (duration/time differences).

ðŸ’  **Usage:** Use when working with time differences or durations

````python
td_idx = pd.TimedeltaIndex(['1 days', '2 days', '3 days'])
df = pd.DataFrame({'progress': [25, 50, 75]}, index=td_idx)
print(df)
````

- **`PeriodIndex`**

```python
pd.PeriodIndex(data=None, freq=None, name=None)
```
â–¶ **Definition** â€“ Index for period values (time spans like months, quarters).

ðŸ’  **Usage:** Use for data that represents time periods rather than specific time points.

````python
periods = pd.period_range('2024-01', periods=4, freq='M')
df = pd.DataFrame({'revenue': [10000, 12000, 15000, 11000]}, index=periods)
print(df)
````

- **`CategoricalIndex`**

```python
pd.CategoricalIndex(data=None, categories=None, ordered=None, name=None)
```
â–¶ **Definition** â€“ Index for categorical data with optional ordering.

ðŸ’  **Usage:** Use when your index has a limited set of categories, especially for memory efficiency.


````python
cat_idx = pd.CategoricalIndex(['small', 'medium', 'large', 'small'],
                              categories=['small', 'medium', 'large'],
                              ordered=True)
df = pd.DataFrame({'price': [10, 20, 30, 15]}, index=cat_idx)
print(df)
````

- **`MultiIndex`**

```python
pd.MultiIndex.from_tuples(tuples, names=None)
pd.MultiIndex.from_arrays(arrays, names=None)
pd.MultiIndex.from_product(iterables, names=None)
```
â–¶ **Definition** â€“ Multi-level or hierarchical index for advanced data organization.

ðŸ’  **Usage:** Use when you need hierarchical indexing for complex data structures.

````python
# From tuples
tuples = [('A', 1), ('A', 2), ('B', 1), ('B', 2)]
multi_idx = pd.MultiIndex.from_tuples(tuples, names=['letter', 'number'])
df = pd.DataFrame({'values': [10, 20, 30, 40]}, index=multi_idx)
print(df)
````

````python
# From arrays
arrays = [['A', 'A', 'B', 'B'], [1, 2, 1, 2]]
multi_idx2 = pd.MultiIndex.from_arrays(arrays, names=['letter', 'number'])
df = pd.DataFrame({'values': [10, 20, 30, 40]}, index=multi_idx2)
print(df)
````

#### **Methods:**

- **`is_unique`**

```python
index.is_unique
```
â–¶ **Definition** â€“ Returns True if the index has no duplicate values.

ðŸ’  **Usage:** Use to check if your index contains unique values only.

````python
idx1 = pd.Index([1, 2, 3, 4])
print(idx1.is_unique)  # True

idx2 = pd.Index([1, 2, 2, 3])
print(idx2.is_unique)  # False
````

- **`equals()`**

```python
index1.equals(index2)
```
â–¶ **Definition** â€“ Checks if two indexes are equal (same values and order).

ðŸ’  **Usage:** Use to compare two indexes for equality.

````python
idx1 = pd.Index([1, 2, 3])
idx2 = pd.Index([1, 2, 3])
idx3 = pd.Index([3, 2, 1])

print(idx1.equals(idx2))  # True
print(idx1.equals(idx3))  # False
````

- **`identical()`**

```python
index1.identical(index2)
```
â–¶ **Definition** â€“ Checks if two indexes are identical (same values, order, and metadata).

ðŸ’  **Usage:** Use when you need to check not just values but also metadata like names.

````python
idx1 = pd.Index([1, 2, 3], name='numbers')
idx2 = pd.Index([1, 2, 3], name='numbers')
idx3 = pd.Index([1, 2, 3], name='values')

print(idx1.identical(idx2))
print(idx1.identical(idx3))
````

- **`union()`**

```python
index1.union(index2, sort=None)
```
â–¶ **Definition** â€“ Returns the union of two indexes (all unique values from both).

ðŸ’  **Usage:** Use to combine indexes and get all unique values.

````python
idx1 = pd.Index([1, 2, 3])
idx2 = pd.Index([3, 4, 5])
union_idx = idx1.union(idx2)
print(union_idx)
````

- **`intersection()`**

```python
index1.intersection(index2, sort=False)
```
â–¶ **Definition** â€“ Returns the intersection of two indexes (common values).

ðŸ’  **Usage:** Use to find common values between two indexes.

````python
idx1 = pd.Index([1, 2, 3, 4])
idx2 = pd.Index([3, 4, 5, 6])
intersection_idx = idx1.intersection(idx2)
print(intersection_idx)
````

- **`difference()`**

```python
index1.difference(index2, sort=None)
```
â–¶ **Definition** â€“ Returns values in the first index that are not in the second.

ðŸ’  **Usage:** Use to find values that exist in one index but not another.

````python
idx1 = pd.Index([1, 2, 3, 4])
idx2 = pd.Index([3, 4, 5, 6])
diff_idx = idx1.difference(idx2)
print(diff_idx)  # Index([1, 2])
````

- **`drop_duplicates()`**

```python
index.drop_duplicates(keep='first')
```
â–¶ **Definition** â€“ Returns index with duplicate values removed.

ðŸ’  **Usage:** Use to remove duplicate values from an index.


```python

```
â–¶ **Definition** â€“ Sets the name(s) of the index.

````python
idx = pd.Index([1, 2, 3, 4])
named_idx = idx.set_names('ID')
print(named_idx.name)
````

````python
# For MultiIndex
multi_idx = pd.MultiIndex.from_tuples([('A', 1), ('B', 2)])
named_multi = multi_idx.set_names(['Letter', 'Number'])
print(named_multi.names)
````

â–¶ **Definition** â€“ Returns the index as a Python list.

ðŸ’  **Usage:** Use when you need to convert the index to a standard Python list.

````python
idx = pd.Index(['a', 'b', 'c', 'd'])
index_list = idx.to_list()
print(index_list)
print(type(index_list))
````

````python
# Create a complex MultiIndex DataFrame
arrays = [['A', 'A', 'B', 'B'], ['X', 'Y', 'X', 'Y']]
multi_idx = pd.MultiIndex.from_arrays(arrays, names=['Group', 'Subgroup'])
df = pd.DataFrame({
    'Value1': [1, 2, 3, 4],
    'Value2': [10, 20, 30, 40]
}, index=multi_idx)

print(df)
print("\nIndex names:", df.index.names)
print("Index levels:", df.index.levels)
print("Is unique:", df.index.is_unique)

# Working with MultiIndex
print("\nGroup A data:")
print(df.loc['A'])

print("\nSpecific cell (A, X):")
print(df.loc[('A', 'X'), 'Value1'])

# Index operations
print("\nUnique values in first level:")
print(df.index.get_level_values(0).unique())

print("\nUnique values in second level:")
print(df.index.get_level_values(1).unique())
````
